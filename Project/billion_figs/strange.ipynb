```python
# Plots the orbital angle phi at each apastron (furthest separation)
spacing = 30
t_apastrons = []
phi_apastrons = []
for n in range(50):
    r_sub = rs[n*spacing:(n+1)*spacing]
    ind = np.where(r_sub==r_sub.max())[0][0] + n*spacing
    t_apastrons.append(times[ind])
    phi_apastrons.append(phis[ind] % (2 * np.pi))

pl.plot(t_apastrons, phi_apastrons, 'x', lw=8)
pl.show()
# Again the weird period behavior
```
