```python
# loads all inspiral data

import numpy as np
import matplotlib.pyplot as pl
import cPickle as pickle
import plot_defaults

####################################
# General Constants (CGS units)
ggrav = 6.67e-8
c = 3.0e10
msun = 1.99e33
rsun = 6.96e10
direcs = [0, 1, 2] # indices corresponding to the three spatial directions

####################################
# Parameters

output_dir = "inspiral1" # name of directory containing simulation output data
start_iteration = 0 # the first iteration to load
n_points = 1e7 # number of iterations in the simulation
input_every = 1000 # inputs one data file for every `input_every` data files
m1 = 1.4*msun # mass of body 1
m2 = 1.4*msun # mass of body 2

####################################
# Load Data

n_loaded = int(n_points / input_every) # number of data files to load

# array of time at each time step
times = np.zeros(n_loaded)

# array of true anomaly at each time step
phis = np.zeros(n_loaded)

# array of semi-major axis at each time step
axs = np.zeros(n_loaded)

# array of eccentricity at each time step
eccs = np.zeros(n_loaded)

# array of separation vector at each time step
xs = np.zeros((3, n_loaded))

# array of reduced quadrupole tensor at each time step
I_bars = np.zeros((3, 3, n_loaded))

# array of plus-polarized gravitational wave strain at each time step
h_pluses = np.zeros(n_loaded)

# array of cross-polarized gravitational wave strain at each time step
h_crosses = np.zeros(n_loaded)

for it in range(n_loaded):
    f = open("{}/iteration-{}.pickle".format(output_dir, start_iteration + it*input_every), 'rb')
    data = pickle.load(f)
    f.close()
    
    times[it], phis[it], axs[it] = data['t'], data['phi'], data['a']
    eccs[it], xs[:,it] = data['e'], data['x']
    h_pluses[it], h_crosses[it] = data['h_plus'], data['h_cross']
    try:
        I_bars[:,:,it] = data['I_bar2']
    except KeyError: # backwards compatibility with old format of output data
        I_bars[:,:,it] = data['I_bar']
    try:
        m1, m2 = data['m1'], data['m2']
    except KeyError:
        pass # use the parameter values defined in this file

####################################
# Analyze Data

M = m1 + m2 # total mass
mu = (m1 * m2) / M # reduced mass

# Calculate z-component of orbital angular momentum
Ls = np.sqrt(ggrav * M * axs * (1 - eccs**2))

# Calculate total mechanical energy of the system
Es = 0.5 * (ggrav * M / Ls)**2 * (eccs**2 - 1)

times1, phis1, axs1, eccs1, xs1, h_pluses1, h_crosses1, Ls1, Es1 = times, phis, axs, eccs, xs, h_pluses, h_crosses, Ls, Es



####################################
# Parameters

output_dir = "inspiral2" # name of directory containing simulation output data
start_iteration = 10000000 # the first iteration to load
n_points = 1e7 # number of iterations in the simulation
input_every = 1000 # inputs one data file for every `input_every` data files
m1 = 1.4*msun # mass of body 1
m2 = 1.4*msun # mass of body 2

####################################
# Load Data

n_loaded = int(n_points / input_every) # number of data files to load

# array of time at each time step
times = np.zeros(n_loaded)

# array of true anomaly at each time step
phis = np.zeros(n_loaded)

# array of semi-major axis at each time step
axs = np.zeros(n_loaded)

# array of eccentricity at each time step
eccs = np.zeros(n_loaded)

# array of separation vector at each time step
xs = np.zeros((3, n_loaded))

# array of reduced quadrupole tensor at each time step
I_bars = np.zeros((3, 3, n_loaded))

# array of plus-polarized gravitational wave strain at each time step
h_pluses = np.zeros(n_loaded)

# array of cross-polarized gravitational wave strain at each time step
h_crosses = np.zeros(n_loaded)

for it in range(n_loaded):
    f = open("{}/iteration-{}.pickle".format(output_dir, start_iteration + it*input_every), 'rb')
    data = pickle.load(f)
    f.close()
    
    times[it], phis[it], axs[it] = data['t'], data['phi'], data['a']
    eccs[it], xs[:,it] = data['e'], data['x']
    h_pluses[it], h_crosses[it] = data['h_plus'], data['h_cross']
    try:
        I_bars[:,:,it] = data['I_bar2']
    except KeyError: # backwards compatibility with old format of output data
        I_bars[:,:,it] = data['I_bar']
    try:
        m1, m2 = data['m1'], data['m2']
    except KeyError:
        pass # use the parameter values defined in this file

####################################
# Analyze Data

M = m1 + m2 # total mass
mu = (m1 * m2) / M # reduced mass

# Calculate z-component of orbital angular momentum
Ls = np.sqrt(ggrav * M * axs * (1 - eccs**2))

# Calculate total mechanical energy of the system
Es = 0.5 * (ggrav * M / Ls)**2 * (eccs**2 - 1)

times2, phis2, axs2, eccs2, xs2, h_pluses2, h_crosses2, Ls2, Es2 = times, phis, axs, eccs, xs, h_pluses, h_crosses, Ls, Es

####################################
# Parameters

output_dir = "inspiral3" # name of directory containing simulation output data
start_iteration = 20000000 # the first iteration to load
n_points = 1e7 # number of iterations in the simulation
input_every = 1000 # inputs one data file for every `input_every` data files
m1 = 1.4*msun # mass of body 1
m2 = 1.4*msun # mass of body 2

####################################
# Load Data

n_loaded = int(n_points / input_every) # number of data files to load

# array of time at each time step
times = np.zeros(n_loaded)

# array of true anomaly at each time step
phis = np.zeros(n_loaded)

# array of semi-major axis at each time step
axs = np.zeros(n_loaded)

# array of eccentricity at each time step
eccs = np.zeros(n_loaded)

# array of separation vector at each time step
xs = np.zeros((3, n_loaded))

# array of reduced quadrupole tensor at each time step
I_bars = np.zeros((3, 3, n_loaded))

# array of plus-polarized gravitational wave strain at each time step
h_pluses = np.zeros(n_loaded)

# array of cross-polarized gravitational wave strain at each time step
h_crosses = np.zeros(n_loaded)

for it in range(n_loaded):
    f = open("{}/iteration-{}.pickle".format(output_dir, start_iteration + it*input_every), 'rb')
    data = pickle.load(f)
    f.close()
    
    times[it], phis[it], axs[it] = data['t'], data['phi'], data['a']
    eccs[it], xs[:,it] = data['e'], data['x']
    h_pluses[it], h_crosses[it] = data['h_plus'], data['h_cross']
    try:
        I_bars[:,:,it] = data['I_bar2']
    except KeyError: # backwards compatibility with old format of output data
        I_bars[:,:,it] = data['I_bar']
    try:
        m1, m2 = data['m1'], data['m2']
    except KeyError:
        pass # use the parameter values defined in this file

####################################
# Analyze Data

M = m1 + m2 # total mass
mu = (m1 * m2) / M # reduced mass

# Calculate z-component of orbital angular momentum
Ls = np.sqrt(ggrav * M * axs * (1 - eccs**2))

# Calculate total mechanical energy of the system
Es = 0.5 * (ggrav * M / Ls)**2 * (eccs**2 - 1)

times3, phis3, axs3, eccs3, xs3, h_pluses3, h_crosses3, Ls3, Es3 = times, phis, axs, eccs, xs, h_pluses, h_crosses, Ls, Es

times = np.hstack((times1, times2, times3))
phis = np.hstack((phis1, phis2, phis3))
axs = np.hstack((axs1, axs2, axs3))
eccs = np.hstack((eccs1, eccs2, eccs3))
xs = np.hstack((xs1, xs2, xs3))
h_pluses = np.hstack((h_pluses1, h_pluses2, h_pluses3))
h_crosses = np.hstack((h_crosses1, h_crosses2, h_crosses3))
Ls = np.hstack((Ls1, Ls2, Ls3))
Es = np.hstack((Es1, Es2, Es3))
```

```python
def ax_analytic(t, a0):
    return (a0**4 - 4 * t * (64./5) * ggrav**3 * m1 * m2 * M / c**5)**0.25
```

```python
fig = pl.figure(figsize=(10,8))
fig.subplots_adjust(left=0.2)
fig.subplots_adjust(bottom=0.16)
fig.subplots_adjust(top=0.85)
fig.subplots_adjust(right=0.85)

pl.xlabel("Time [s]")
pl.ylabel(r"Semi-major axis [$\times 10^7$ cm]")
pl.xlim(times[0] - 162.60404, times[-1] - 162.60404)

ax = pl.gca()
xmajorLocator = pl.MultipleLocator(50)
ax.xaxis.set_major_locator(xmajorLocator)

numer, = pl.plot(times - 162.60404, axs/1e7, lw=8)
analyt, = pl.plot(times - 162.60404, ax_analytic(times, 4.6e7)/1e7, 'r--', lw=8)

pl.legend( (numer, analyt), ("Numerical Result", "Analytic Result"), frameon=False, loc="lower left")

pl.savefig("inspiral_figs/inspiral_axs.pdf")
pl.show()
```
